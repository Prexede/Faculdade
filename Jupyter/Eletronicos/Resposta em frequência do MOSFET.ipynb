```python
import numpy as np
import sympy
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

## Componentes

```python
K = 10**3
u = 10**-6
m = 10**-3
M = 10**6

## REAL
#Resistores
Rg1 = 330*K
Rg2 = 100*K
Rs = 39
Rl = 10*K
Rd = 1*K
Rp = 10

#Capacitores
Cin = 1*u
Co = 1*u

#OUTROS
Vcc = 18
Vt = 3.788
k = 3.42045
```

```python
## CALCULO ID
Id1  = sympy.symbols('Id')
expr = k*(((((Rg2*Vcc)/(Rg1+Rg2))-(Id1*Rs))-Vt)**2) - Id
solucao1 = sympy.solve(expr, Id1)
Id1 = solucao1[0]

print("CIRCUITO 1")
print("IDs:",solucao1)

## Auxiliar 
Req2 = (Rs*Rp)/(Rs+Rp)

Id2  = sympy.symbols('Id')
expr = k*(((((Rg2*Vcc)/(Rg1+Rg2))-(Id2*Req2))-Vt)**2) - Id2
solucao2 = sympy.solve(expr, Id2)
Id2 = solucao2[0]
print("CIRCUITO 2")
print("IDs:",solucao2)
```

```python
## CIRCUITO 1 
Vd1 = Vcc - (Rd*Id1)
Vs1 = Rs*Id1
Vg1 = Rg2*Vcc/(Rg1+Rg2)

Vds1 = Vd1 - Vs1
Vgs1 = Vg1 - Vs1


## CIRCUITO 2
Vd2 = Vcc - (Rd*Id2)
Vs2 = Req2*Id2
Vg2 = Rg2*Vcc/(Rg1+Rg2)

Vds2 = Vd2 - Vs2
Vgs2 = Vg2 - Vs2


print("CIRCUITO 1")
print("Vds:",Vds1,"Vgs:",Vgs1)

print("CIRCUITO 2")
print("Vds:",Vds2,"Vgs:",Vgs2)
```

```python
Gm = 2*k*(Vgs - Vt)
Req = (Rd*Rl)/(Rd+Rl)


Av1 = -((Gm)*Req)/(1+(Gm*Rs))
Av2 = -((Gm)*Req)/(1+(Gm*Req2))

AvL1 = 20*math.log(-Av1,10)
AvL2 = 20*math.log(-Av2,10)

## CIRCUITO 1
print("CIRCUITO 1")
print("Valor de Gm:",Gm,"\nGanho Av:",Av1,"\nGanho em Db:",AvL1)

print("\nCIRCUITO 2")
## CIRCUITO 2
print("Valor de Gm:",Gm,"\nGanho Av:",Av2,"\nGanho em Db:",AvL2)
```
