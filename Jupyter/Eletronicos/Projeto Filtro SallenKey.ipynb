# Bibliotecas

```python
import numpy as np
```

## Especificações

```python
## EXPOENTES
n = 10**-9

## Fatores 
a1 = 1.065
b1 = 1.9305

#Componentes
C1 = 22*n

## Frequencia de corte
Fc = 9877
```

## Calculo dos Componentes

```python
C2 = (C1*4*b1)/(a1**2)

R1 = ((a1*C2)+(np.sqrt(((a1**2)*(C2**2)) - (4*b1*C1*C2)))) / (4*np.pi*Fc*C1*C2)
R2 = ((a1*C2)-(np.sqrt(((a1**2)*(C2**2)) - (4*b1*C1*C2)))) / (4*np.pi*Fc*C1*C2)

print("Valores encontrados:","\nResistor 1[ohms]:",R1,"\nResistor 2[ohms]:",R2,"\nCapacitor 2[nF]:",round(C2,15))
```
