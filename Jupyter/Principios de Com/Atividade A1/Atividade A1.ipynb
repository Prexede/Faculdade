```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

#Definicao de Variaveis
t = np.linspace(0,0.6,500)

#1a
w1 = 10

#1b
w2 = 20

#1c
w3 = 100

#Plotando Grafico

figure(figsize=(8, 6), dpi=100)

plt.plot(t, np.sin(w1*t),'r')
plt.plot(t, np.sin(w2*t),'b')
plt.plot(t, np.sin(w3*t),'g')


plt.title('Funçao Seno')
plt.xlabel('X', fontsize=15)
plt.ylabel('F(X)', fontsize=15)
plt.legend([10,20,100])

#Separado
figure(figsize=(10, 8), dpi=120)
fig, (ax1,ax2,ax3) = plt.subplots(3,dpi = 100)

ax1.plot(t, np.sin(w1*t),'r')
ax1.legend([10])
ax2.plot(t, np.sin(w2*t),'b')
ax2.legend([20])
ax3.plot(t, np.sin(w3*t),'g')
ax3.legend([100])
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

#Definicao de Variaveis
t = np.linspace(0,1,500)

#2a
Ft1 = np.sin(10*t) + np.sin(20*t)

#2b
Ft2 = np.sin(10*t) + np.sin(20*t) + np.sin(30*t)

#Separado
figure(figsize=(10, 8), dpi=120)
fig, (ax1,ax2) = plt.subplots(2,dpi = 100)

ax1.plot(t, Ft1,'r')
ax1.legend("a")
ax2.plot(t, Ft2,'b')
ax2.legend("b")
```
