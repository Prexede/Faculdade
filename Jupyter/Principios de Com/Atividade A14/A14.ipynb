# Bibliotecas

```python
import numpy as np
import scipy.fftpack
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scipy.fftpack import fft, fftfreq
```

# Sinais Fundamentais

```python
### parâmetros do sinal ###

##Amplitude 
A = 1
A1 = 1

##Frequencia 
##Portadora
freq = 5000
##Modulante
freq1 = 500

##Fator BETA
B = 10

##Parametros de Sinal
Amostras_ciclo = 100

#Tempo inicial
t0 = 0

#Qtd de Periodos
Num_T = 100

#Periodo
T = 1/freq1

## N 
N = Num_T*Amostras_ciclo

#Fator de Modulação
#sobremodulado
u = 1.5
#normal
u1 = 0.5

##Vetor de tempo
t = np.linspace(t0,T*Num_T, Num_T*Amostras_ciclo)

# sinal portador
#w = freq
w = 2.0*np.pi*freq
X = A*np.cos(w*t)

# sinal modulante
#w1= freq1
w1 = 2.0*np.pi*freq1

# Sinal FM
Y = A1*np.cos(2*np.pi*freq*t - (B*(np.cos(2*np.pi*freq1*t))))

# Sinal Portador
Z = A*np.cos(2*np.pi*freq*t);

# Sinal Modulante
K = A1*np.cos(2*np.pi*freq1*t);
```

## Graficos

```python
figure(figsize=(15, 5), dpi=100)
plt.plot(t,Z,'r')
plt.xlim(0,0.05)
plt.xlabel('t(s)')
plt.ylabel('Amplitude')
plt.title('Sinal Portador')
plt.show()

figure(figsize=(15, 5), dpi=100)
plt.plot(t,Y,'g')
plt.xlim(0,0.005)
plt.xlabel('t(s)')
plt.ylabel('Amplitude')
plt.title('Sinal Modulado em FM')
plt.show()

figure(figsize=(15, 5), dpi=100)
plt.plot(t,K,'b')
plt.xlim(0,0.005)
plt.xlabel('t(s)')
plt.ylabel('Amplitude')
plt.title('Sinal Modulante')
plt.show()
```

## Transformada de Fourier

```python
zf = scipy.fftpack.fft(Y)
#xf = np.linspace(0.0, T/Amostras_ciclo, N//2)
xf = np.linspace(0.0, 1.0//(2.0*T), N//2)

figure(figsize=(15, 5), dpi=100)
#plt.xlim([0,200])
#plt.ylim([0,0.4])
plt.grid()

plt.title('Sinal da Modulante dominio frequência')
plt.xlabel("Frequência (Hz)")
plt.ylabel("Amplitude")
#plt.plot(xf, np.abs(zf[0:N//2]),'r')
plt.plot(xf, 2.0/N * np.abs(zf[:N//2]),'r')

plt.show()
```
