# Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# Sinais Fundamentais

```python
### parâmetros do sinal ###

##Amplitude 
A = 1
A1 = 1

##Frequencia 
freq = 50
freq1 = 5

##Parametros de Sinal
Amostras_ciclo = 1000 

#Tempo inicial
t0 = 0

#Qtd de Periodos
Num_T = 10

#Periodo
T = 1/freq1

#Fator de Modulação
#sobremodulado
u = 1.5
#normal
u1 = 0.5

##Vetor de tempo
t = np.linspace(t0,T*Num_T, Num_T*Amostras_ciclo)

# sinal portador
#w = 2.0*np.pi*freq
w = freq
X = A*np.cos(w*t)

# sinal modulante
#w1 = 2.0*np.pi*freq1
w1= freq1
Y = A1*np.cos(w1*t)

#sinal modulado 
Z = A*(1+(u*np.cos(w1*t)))*np.cos(w*t)
```

## Plotando os Graficos 

### Sinal Portador 

```python
##Tamanho da Figura 
figure(figsize=(15, 5), dpi=100)

##Cabeçalho
plt.title('Sinal portador')
plt.xlabel('X', fontsize=15)
plt.ylabel('F(X)', fontsize=15)
plt.grid()

##plot linhas traçejadas 
plt.plot([t0,T*Num_T], [0,0], '--')

##plot função 
plt.plot(t,X ,'g--')

##Salvando valores 
plt.savefig(r'C:\Users\carlo\OneDrive\Documentos\GitHub\Faculdade\Principios de Com\Atividade A4\portador.png', dpi = 300)
```

## Sinal Modulante

```python
##Tamanho da Figura 
figure(figsize=(15, 5), dpi=100)

##Cabeçalho
plt.title('Sinal modulante')
plt.xlabel('X', fontsize=15)
plt.ylabel('F(X)', fontsize=15)
plt.grid()

##plot linhas traçejadas 
plt.plot([t0,T*Num_T], [0,0], '--')

##plot função 
plt.plot(t,Y ,'b:')

##Salvando valores 
plt.savefig(r'C:\Users\carlo\OneDrive\Documentos\GitHub\Faculdade\Principios de Com\Atividade A4\modulante.png', dpi = 300)
```

## Sinal Modulado

```python
##Tamanho da Figura 
figure(figsize=(15, 5), dpi=100)

##Cabeçalho
plt.title('Funçao Modulada')
plt.xlabel('X', fontsize=15)
plt.ylabel('F(X)', fontsize=15)
plt.grid()

##plot linhas traçejadas 
plt.plot([t0,T*Num_T], [0,0], '--')

##plot função 
plt.plot(t,Z,'r')

##Salvando valores 
plt.savefig(r'C:\Users\carlo\OneDrive\Documentos\GitHub\Faculdade\Principios de Com\Atividade A4\modulado0.png', dpi = 300)
```
