```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# 1Hz sine wave
npts  = 100
tmax  = 10                          
t     = np.linspace(0, tmax, npts, endpoint=False)
x     = 1/2*np.pi
y     = np.sin(2*np.pi*x*t)
dt    = tmax/npts

# FFT computation
Y      = fft(y)
freq   = fftfreq(npts, d=dt)
amplit = abs(Y)/npts
phase  = np.angle(Y)
phase  = phase / np.pi

fig, ax = plt.subplots(1, 2, figsize=(10,4))
ax[0].plot(freq, amplit)
ax[1].plot(freq, phase)

ax[0].plot([2,2], [0,0.5], '--')
ax[0].plot([-2,-2], [0,0.5], '--')
ax[1].plot([2,2], [-1,1], '--')
ax[1].plot([-2,-2], [-1,1], '--')
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 100
dt = 1 / fs
t = np.arange(0, 10, dt)
F = 2*np.pi

y = np.cos(2*np.pi*F*t)
S = np.fft.fftshift(np.fft.fft(y) / y.shape[0])

f_scale = np.linspace(-1, 1, y.shape[0]) * (fs / 2)


a = np.abs(S)
phi = np.angle(S)

plt.subplot(2, 1, 1, title="amplitude")
plt.plot(f_scale, a)  
plt.title("Diagrama de Espectro de magnitude") 
plt.show() 


plt.subplot(2, 1, 2, title="phase")
plt.plot(f_scale, phi)
plt.title("Diagrama de Espectro de Fase") 
plt.show() 

plt.show()
```
